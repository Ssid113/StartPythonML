1.__Реализовать игру__: Камень-Ножницы-Бумага-Ящерица-Спок
__Правила игры__: Ножницы разрезают бумагу. Бумага накрывает камень. Камень бьет по ящерице. Ящерица выпускает яд в Спока. Спок разбивает ножницы. Ножницы отрезают голову ящерице. Ящерица съедает бумагу. Бумага опровергает Спока. Спок превращает камень в пар. И, как всегда, камень бьет по ножницам.
__Действия игрока__: Ввести свое имя, ввести колличество игроков, выбрать жест(Камень-Ножницы-Бумага-Ящерица-Спок), наслаждаться игрой.
__Дополнительные условия__: в игре могут учавствовать от 2-х до 9 человек. Вы должны играть против ботов. Игра на выбывание, победитель определяется когда остается только один, при ничье раунд повторяется. 
__После чего вывести статистику игры в виде__:
Раунд 1    | вылетел игрок1, игрок5
Раунд 2    | ничья
Раунд 3    | вылетел игрок 2, игрок3
Победитель | игрок4
